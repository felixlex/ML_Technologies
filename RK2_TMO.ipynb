## Импортируем все необходимые библиотеки

### Производим кодирование категориальных данных

### Разделим выборку на обучающую и тестовую

```python
df.loc[:,'MarriedID':'PositionID']
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.loc[:,'MarriedID':'PositionID'], df["PerformanceScore"], test_size=0.2, random_state=42)
```

## Решим задачу классификации при помощи метода логистической регрессии

В отличие от обычной регрессии, в методе логистической регрессии не производится предсказание значения числовой переменной исходя из выборки исходных значений. Вместо этого, значением функции является вероятность того, что данное исходное значение принадлежит к определенному классу.

```python
clf_lin = LogisticRegression()
clf_lin.fit(X_train,y_train)
```

## Градиентный бустинг

```python
clf_gb = GradientBoostingClassifier()
clf_gb.fit(X_train,y_train)
```

## Оценим качество моделей

```python
# Metrics
clfs = [clf_lin,clf_gb]
for clf in clfs:
    print(f"{type(clf).__name__}:")
    print('\t',f"accuracy = {accuracy_score(y_test,clf.predict(X_test))}")
    print('\t',f"f1 = {f1_score(y_test,clf.predict(X_test),average='micro')}")
```

Оценив качество моделей, мы можем заметить, что градиентный бустинг показывает идеальное значение для обеих метрик, что означает, что модель идеально предсказывает классы. С другой стороны, модель логистической регрессии тоже показывает неплохие результаты близкие к единице.

Таким образом, на основании предоставленных результатов можно сделать вывод, что модель GradientBoostingClassifier предпочтительнее для этого набора данных, так как она обеспечивает более высокую точность и F1-меру по сравнению с моделью LogisticRegression.
